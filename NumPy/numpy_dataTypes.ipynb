```python
#Data types
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr, arr.dtype)
```

```python
arr2 = np.array([1, 2, 3.14, 4, 5]) #if one of the values is float, every value will be float
print(arr2, arr2.dtype)
```

```python
arr3 = np.array([2,7, 9, "number"]) #Unicode string 
print(arr3, arr3.dtype)
```

```python
arr4 = np.array([5 + 3j]) #complex number is combination of two float64, hence complex128
print(arr4, arr4.dtype)
```

```python
arr5 = np.array(["prime", {1, 2, 3}, 3.14])
print(arr5, arr5.dtype)
```
